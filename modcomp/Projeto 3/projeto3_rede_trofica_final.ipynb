Para o projeto 3 da disciplina de Modelagem Computacional, o grupo teve por objetivo compreender e modelar um exemplo de uma rede trófica, relacionando recursos naturais de vegetação, espécies de carnívoros e herbívoros, bem como também realizar diversas simulações com diversos parâmetros a fim de reproduzir diferentes tipos de situações do ambiente em questão.

A rede trófica contém relações próprias entre os seguintes seres vivos:

* Planta (produtor da rede);
* Formiga;
* Tatu;
* Tamanduá;
* Pardal;
* Onça;
* Jacaré.

rede_trofica_diagrama2.png

Foram utilizadas as seguintes equações para descrever algebricamente o comportamento de cada uma das populações em relação ao tempo:

* $                         
\frac{dP}{dt} = P*(p1 - (p1 * P / K) - p2 * F - p3 *TATU)
$

* $
\frac{dF}{dt} = F*(-p4 + p5*P - p6*PAR - p7*TAM) 
$                     

* $
\frac{dTATU}{dt} = TATU*(- p8 + p9*P - p10*JAC - p11*ONC)
$

* $
\frac{dPAR}{dt} = PAR*(- p12 + p13*F - p14*JAC)
$

* $
\frac{dJAC}{dt} = JAC*(- p15 + p16*PAR + p17*TATU)
$

* $
\frac{dTAM}{dt} = TAM*(- p18 + p19*F - p20*ONC)*dt
$

* $
\frac{dONC}{dt} = ONC*(- p21 + p22*TAM + p23*TATU)
$


Onde: 

* p1,...,p23 = paramêtros das equações;
* K = grau de saturação do ambiente;
* P = população de plantas;
* F = população de formigas;
* TATU = população de tatus;
* PAR = população de pardais;
* JAC = população de jacarés;
* TAM = população de tamanduás;
* ONC = população de onças;
* dt = variação mínima de tempo (aqui consideraremos igual a 0.01).

#Simulação 1: cálculos e resultados

```python
t = 0 #tempo
dt = 0.01 #variacao do tempo
t_max = 100 #tempo maximo
K = 50 #Capacidade de suporte do ambiente (saturacao da planta)
P = 8 #Populacao Planta
F = 4 #Populacao Formiga
TATU = 4 #Populacao Tatu
PAR = 2 #Populacao Pardal
JAC = 1 # Populacao Jacare
TAM = 2 #Populacao Tamandua
ONC = 1 #Populacao Onca

#listas
lista_t = [] #lista para tempo
lista_P = [] #lista pop planta
lista_F = [] #lista pop formiga
lista_TATU = [] #lista pop tatu
lista_PAR = [] #lista pop pardal
lista_JAC = [] #lista pop jacare
lista_TAM = [] #lista pop tamandua
lista_ONC = [] #lista pop onca

#Loop
while(t<=t_max):
    #Equacoes
  P += P*(p1 - ((p1*P)/K) - p2*F - p3*TATU)*dt #Planta~
  F += F*(- p4 + p5*P - p6*PAR - p7*TAM)*dt #Formiga
  TATU += TATU*(- p8 + p9*P - p10*JAC - p11*ONC)*dt #Tatu
  PAR += PAR*(- p12 + p13*F - p14*JAC)*dt #Pardal
  JAC += JAC*(- p15 + p16*PAR + p17*TATU)*dt #Jacare
  TAM += TAM*(- p18 + p19*F - p20*ONC)*dt #Tamandua
  ONC += ONC*(- p21 + p22*TAM + p23*TATU)*dt #Onca

    #Listas recebendo valores
  lista_P.append(P) #Planta
  lista_F.append(F) #Formiga
  lista_TATU.append(TATU) #Tatu
  lista_PAR.append(PAR) #Pardal
  lista_JAC.append(JAC) #Jacare
  lista_TAM.append(TAM) #Tamandua
  lista_ONC.append(ONC) #Onca
  lista_t.append(t) #Tempo

    #Acumulo da variacao de tempo
  t += dt

#Graficos
plt.plot(lista_t, lista_P, color='green')
plt.plot(lista_t, lista_F, color='red')
plt.plot(lista_t,lista_TAM, color='blue')
plt.plot(lista_t, lista_ONC, color='orange')
plt.plot(lista_t, lista_JAC, color='yellow')
plt.plot(lista_t, lista_TATU, color='purple')
plt.plot(lista_t, lista_PAR, color='black')
plt.plot()

plt.xlabel("Tempo")
plt.ylabel("Populacao (valores arbitrários)")
plt.show()
```

O gráfico acima representa o sistema sem pertubações e equilíbrio, ou seja, nenhuma ser vivo é extinto nesse período de tempo. Nele, é possível observar as variações de população das espécies em função do tempo. Cada ser vivo é representado por uma cor de linha:

*   Planta: verde;
*   Formiga: vermelha;
*   Tatu: roxa;
*   Pardal: preta;
*   Tamanduá: azul;
