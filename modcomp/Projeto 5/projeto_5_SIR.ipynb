As simulações apresentadas neste trabalho salientam uma situação fictícia de uma possível epidemia de gripe em uma determinada região da América do Norte, possibilitando observar comportamento de cada tipo de população ao longo das estações do ano.





##Equações

Parâmetros:
* $dt =$ Taxa de variação do tempo (padrão igual a 0.01)
* $r =$ Taxa de infecção
* $a =$ Taxa de recuperação

Populações:
* $S =$ População de indivíduos suscetíveis
* $I =$ População de indivíduos infectados
* $R =$ População de indivíduos recuperados
* $N = S + I + R =$ População total

Leis que descrevem o comportamento das populações S, I, R:
* $\frac{dS}{dt} = -rSI$

* $\frac{dI}{dt} = rSI - aI$

* $\frac{dR}{dt} = aI$






##Simulações

```python
#Bibliotecas utilziadas
import math
import numpy as np
from matplotlib import pyplot as plt
```

#Situação 1
Numa temporada de outono/inverno, isto é, entre os meses de setembro e março, o grupo definiu uma situação de frio extremo sem a obrigação de isolamento social, onde taxa de infecção 20x maior que a taxa de recuperação. A população total foi definida como 1, onde os indivíduos suscetíveis, infectados e recuperados representam, respectivamente, 95%, 5% e 0% da população no tempo inicial $t=$0.

Parâmetros:
* $dt = $ 0.01
* $r =$ 0.04
* $a =$ 0.002

Populações:
* $S =$ 0.95
* $I =$ 0.05
* $R =$ 0
* $N =$ 1

```python
#SITUAÇÃO 1

#Tempo
t = 0 #Tempo inicial
dt = 0.01 #Variação de Tempo
t_max = 3000 #Tempo máximo
lista_t = []

#Parâmetros
r = 0.04 #Taxa de infecção
a = 0.002 #Taxa de recuperação

#Populações
S = 0.95 #População Suscetíveis
I = 0.05 #População Infectados
R = 0 #População Recuperados
lista_S = []
lista_I= []
lista_R = []
lista_N = []

#Loop
while(t<t_max):
  #Equações
  S += (-r*S*I)*dt #Suscetíveis
  I += ((r*S*I)-(a*I))*dt #Infectados
  R += (a*I)*dt #Recuperados
  N = S + I + R #População total

  #Listas recebendo valores
  lista_S.append(S)
  lista_I.append(I)
  lista_R.append(R)
  lista_N.append(N)
  lista_t.append(t)

  t += dt #Acumulo da variação de tempo

#Gráfico
plt.plot(lista_t, lista_N, color='grey', label="População Total (constante)")
plt.plot(lista_t, lista_S, color='blue', label="População de Suscetíveis")
plt.plot(lista_t, lista_I, color='red', label="População de Infectados")
plt.plot(lista_t, lista_R, color='green', label="População de Recuperados")
plt.ylabel("População (em %)")
plt.xlabel("Tempo (em horas)")
plt.title('Comportamento das populações - Situação 1')
plt.legend()
plt.show()
```

No gráfico, é possível observar a rápida acentuação da curva da população infectada, onde o pico chega perto de 80% da população total pouco depois de $t=$250. Nessa situação, todos os indivíduos foram infectados em algum momento, pois a população de suscetíveis é zerada rapidamente. A situação se assemelha a de uma epidemia.

Portanto nessa situação é possível especular que o vírus não teve espaço para infectar novos indíviduos visto que a taxa de infecção é igual a taxa de recuperação.

##Conclusão

A partir das simulações realizadas, é possível inferir que o modelo SIR é uma boa base para o começo de modelagem de sistemas complexos, podendo envolver tipos de transmissão, propagação e evolução de diversos tipos de doenças, em especial em cenários ideais como, por exemplo, o caso de escolas, ou empresas.

Porém, logo de cara, é possível identificar as principais limitações desse modelo, já que além da população necessariamente precisar ser constante ao longo do tempo, ela tem distribuição homogênea no espaço, algo que particularmente, quando se analisa cenários reais, não é possível nem mesmo quando é usado como objeto de estudo um bairro de uma cidade selecionada ao acaso.

Portanto, para além de suas limitações, o SIR é um modelo interessante para se estudar situações específicas de evolução epidemiológicas em espaços pequenos, com baixa variação da população ao longo do tempo. Nesse tipo de cenário, consegue-se um comportamento de desenvolvimento e evolução de doenças semelhantes ao que ocorre na realidade.
